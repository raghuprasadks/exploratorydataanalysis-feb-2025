# Bar chart of Top 5 covid affected countries

```python
df_sorted = df.sort_values(by='Confirmed', ascending=False)
df_top5 = df_sorted.head(5)
df_top5
```

```python
df_stacked = df_top5[['Country/Region', 'Confirmed', 'Deaths', 'Recovered']]
df_stacked
```

```python
df_melted = pd.melt(df_stacked, id_vars='Country/Region', var_name='Status', value_name='Cases')
df_melted
```

```python
df_sorted = df.sort_values(by='Confirmed', ascending=False)
df_top5 = df_sorted.head(5)
df_stacked = df_top5[['Country/Region', 'Confirmed', 'Deaths', 'Recovered']]
df_melted = pd.melt(df_stacked, id_vars='Country/Region', var_name='Status', value_name='Cases')

plt.figure(figsize=(10, 6))
sns.barplot(x='Country/Region', y='Cases', hue='Status', data=df_melted, palette='viridis')
plt.xlabel('Country/Region')
plt.ylabel('Number of Cases')
plt.title('Top 5 COVID-19 Cases by Country/Region (Confirmed, Deaths, Recovered)')
plt.xticks(rotation=45)
plt.legend(title='Status', loc='upper right')
plt.show()
```

# Pie chart of Least Affected countries by covid 19

```python
df_least_affected = df.sort_values(by='Confirmed', ascending=True).head(5)
plt.figure(figsize=(5, 5))
plt.pie(df_least_affected['Confirmed'], labels=df_least_affected['Country/Region'], autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Least Affected Countries by COVID-19 (Confirmed Cases)')
plt.show()
```

# Bar chart for top 8 countries with the most deaths

```python

df_top8_deaths = df.sort_values(by='Deaths', ascending=False).head(8)

# Print the top 8 countries with the most deaths
top8_countries_deaths = df_top8_deaths[['Country/Region', 'Deaths']]
print("Top 8 countries with the most deaths:")
print(top8_countries_deaths)

plt.figure(figsize=(12, 6))
plt.bar(df_top8_deaths['Country/Region'], df_top8_deaths['Deaths'], color='pink')
plt.xlabel('Country/Region')
plt.ylabel('Number of Deaths')
plt.title('Top 8 Countries with the Most COVID-19 Deaths')
plt.xticks(rotation=45)
plt.show()
```

# Heatmap

```python
numeric_df = df.select_dtypes(include=['number'])
correlation_matrix = numeric_df.corr()

plt.figure(figsize=(8, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Numeric Columns')
plt.show()
```

# Donut chart

```python
df['Recovery Rate'] = (df['Recovered'] / df['Confirmed']) * 100


df_top5_recovery = df.sort_values(by='Recovery Rate', ascending=False).head(5)


top5_countries = df_top5_recovery[['Country/Region', 'Recovery Rate']]
print("Top 5 countries with the fastest recovery rates:")
print(top5_countries)

plt.figure(figsize=(8, 5))


plt.pie(df_top5_recovery['Recovery Rate'], labels=df_top5_recovery['Country/Region'], autopct='%1.1f%%',
        wedgeprops=dict(width=0.4), colors=['blue', 'green', 'orange', 'red', 'purple'])

plt.title('Top 5 Countries with the Fastest COVID-19 Recovery Rates')
plt.show()
```
