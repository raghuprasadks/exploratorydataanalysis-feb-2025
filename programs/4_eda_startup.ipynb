```python
import re
def extract_numbers(valuation):
    # Use regular expression to find all numbers in the string
    numbers = re.findall(r'\d+', valuation)

    # If numbers are found, join them and convert to an integer
    if numbers:
        return int(''.join(numbers))
    else:
        return None  # Handle cases with no numbers
```

```python
startupdf['Current Valuation'] = startupdf['Current Valuation'].apply(extract_numbers)
```
