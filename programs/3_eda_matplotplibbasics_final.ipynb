```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
print("matplot lib version ",matplotlib.__version__)
```

```python
x=np.array(["USA","China","Japan","Gemany","India"])
```

```python
x
```

```python
y=np.array([28,18,5,4,3])
```

```python
y
```

```python
print("line chart")
plt.title("Top 5 economies in the world")
plt.xlabel("Country")
plt.ylabel("GDP in USD Trillions")
plt.plot(x,y)
```

```python
print("scatter plot")
plt.title("Top 5 economies in the world")
plt.xlabel("Country")
plt.ylabel("GDP in USD Trillions")
plt.scatter(x,y)
```

```python
print("bar chart - vertical")
plt.title("Top 5 economies in the world")
plt.xlabel("Country")
plt.ylabel("GDP in USD Trillions")
plt.bar(x,y)
```

```python
print("bar chart - horizontal")
plt.title("Top 5 economies in the world")
plt.xlabel("Country")
plt.ylabel("GDP in USD Trillions")
plt.barh(x,y)
```

```python
print("pie chart ")
plt.title("Top 5 economies in the world")
plt.xlabel("Country")
plt.ylabel("GDP in USD Trillions")
plt.pie(y,labels=x)
```
